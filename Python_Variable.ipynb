A Python Variable is a pointer , the pointer consists of the memory block identifier of the memory block which has the value to be ascribed to the variable . Hence , unlike in C which uses value-type variables in Python changing the value of the variable will simply change the pointer to 'point' to the memory block containing the new value . Two immutable variables with same value will have identical pointers in python .

In case of Immutable variables : 

```
var1 = 20
var2 = 20
var3 = 30 
print(id(var1),id(var2),id(var3))
print(var1 is var2)
var2 = 30
print(var2 is var3,var1 is var2)
```

In cases of mutable variables :

```
l1 = [1.32,20,'cod']
l2 = [1.32,20,'cod']
l3=[[1.32,20,'cod'],[13,12],["oop"]]
print(l1[1] is var1)
print(l1 is l2)
print(l1 is l3[0] )
```

```
l1 = [1.32,20,'cod']
l2 = [1.32,20,'cod']
l3=[[1.32,20,'cod'],[13,12],["oop"]]
l4=[[1.32,20,'cod'],l2,l3]
print(l4[0] is l1)
print(l4[1] is l2)
print(l4[2][0],l4[2][0] is l2)
```
