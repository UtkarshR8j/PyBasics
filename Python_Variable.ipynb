A Python Variable is a pointer , the pointer consists of the memory block identifier of the memory block which has the value to be ascribed to the variable . Hence , two variables with same value will have identical pointers in python . Changing the value of the variable will simply change the pointer to 'point' to the memory block containing the new value .
