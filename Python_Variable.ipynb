### Basic Mechanism of variables and  referencing :
Python consists of the memory block which has the value and the attributes i.e. to be ascribed to the new object being created .The variable hence is a short-hand notation of that memory block which has the object and has a unique identification , id . Upon assignment the variable name  'a' is created which is generic in nature and can point to any object in those memory blocks .  Hence , unlike in C which uses value-type variables in Python changing the value of the variable will simply change the pointer to 'point' to the memory block containing the new onject with the required value . 


```
a=3.141
print(a,type(a))
a= 'L'
print(a,type(a))
```

image.png


### Memory reclaimation : 
In the above code a variable name 'a' is pointing to an object (float,3.141) .
When the variable 'a' is coded to refer to a new object (str,'L') , 'a' is assigned a new identity using a id pointing to a memory block containing the new object . 
The older block containing (float,3.141) is reclaimed as in it is emptied a added back to the memory pool , as the variable counter which tracks the number of variables referencing that particular memory block drops to 0 .

### Catching : 

Two immutable variables however with same value will have identical pointers in python .

## In case of Immutable(catched) variables : 

```
var1 = 20
var2 = 20
var3 = 30 
print(id(var1),id(var2),id(var3))
print(var1 is var2)
var2 = 30
print(var2 is var3,var1 is var2)
```

## In cases of mutable variables :

```
l1=[12,"L",19.3]
l2=l1
print(l2)
l1[0]=24
l1[1]=1
print(l2)   # the value of l2 also changes 
print(l2 is l1 , id(l1))
l1=[12,14,15]
print(l2)
print(l2 is l1 , id(l1))
```

- In the above example , initially , $l1$ and $l2$ point to same object , in $140549513477808$ , changing the value of one or more element of $l1$ individually , does not create a new object in a different memory block ,rather it changes the object in this memory block itself and hence change $l2$ also in the process . 

- But if we change the list itself , this does'nt happen . It then uses the earlier pathway of changing the memory block and object .
